```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('people.csv')
ages = data.age
genders = data.gender
print(ages)
print(genders)
```

```python
fig, ax = plt.subplots()
ax.set_title('My people')
ax.set_xlabel("#")
ax.set_ylabel("Age")
colors = [('r' if g == 'f' else 'b') for g in genders]
plt.bar(range(1, len(ages) + 1), ages, color = colors)
plt.show()
```
