```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv('tested.csv')
```

Задание 1.
Распределение выживших среди мужчин и женщин по классу билета.

```python
grouped_by_sex = data[(data.Survived == 1)].groupby(['Sex', 'Pclass']).count().PassengerId
female = grouped_by_sex.loc['female']
# male = grouped_by_sex.loc['male']
```

```python
fig, ax = plt.subplots()
ax.set_title('Class distr (f)')
ax.set_xlabel("Class")
ax.set_ylabel("Count")
ax.set_xticks(female.keys())
ax.bar(female.keys(), female.values)
fig.set_figwidth(10)
plt.show()
```

Задание 2.
Построить круговую диаграмму для признака "класс пассажира" (количество людей в каждом классе)

```python
grouped_by_class = data.groupby("Pclass").count().PassengerId
labels = grouped_by_class.keys()
plt.pie(grouped_by_class.values, labels = labels, autopct='%1.1f%%')
plt.show()
```

Задание 3.
Построить распределение возрастов всех пассажиров.

```python
ages = data[(data.Age.notnull())].Age.map(lambda x: math.floor(x))
print(ages.min())
print(ages.max())
```

```python
fig, ax = plt.subplots()
ax.set_title('Age distr')
ax.set_xlabel('Age')
ax.set_ylabel('Count')
plt.hist(ages, range=(1, 80), bins=80)
plt.show()
```
